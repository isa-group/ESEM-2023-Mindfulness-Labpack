## Data Wrangling

### MAAS

```python
from numpy import nan
# Computing the total sum of responses per experimental subject:
columns=list(maas.columns)
maas['attention awareness'] = maas[columns[3:18]].sum(axis=1)
maas['attention awareness']= maas['attention awareness'].replace(0,nan)
# Renaming conlumns for translation from Spanish:
translatedColumns = {'GRUPO': 'GROUP',
                     'MOMENTO': 'MOMENT',
                     'N Empleado': 'ID'}

maas=maas.rename(columns=translatedColumns)
maas=maas.dropna()
groupModification={'G1':'EXP','G2':'CTRL'}
maas=maas.replace(groupModification)
maas
```

### Welfare

```python
welfare=welfare.rename(columns=translatedColumns)
welfare=welfare.rename(columns={'Grupo':'GROUP'})
#groupModification={'G1':'EXP','G2':'CTRL'}
welfare=welfare.replace(groupModification)
welfare=welfare.dropna()
welfare
```

### Techno-stress (Additional Questions)

```python
columns=list(additionalQuestions.columns)
additionalQuestions['Techno-stress'] = additionalQuestions[columns[3:10]].sum(axis=1)
additionalQuestions['Techno-stress'] = additionalQuestions['Techno-stress'].replace(0,nan)
# We change the direction of the techno-stress to get more meaningful figures:
additionalQuestions['Techno-stress'] = 42 - additionalQuestions['Techno-stress'].replace(0,nan)

 # Column Translation:
additionalQuestions=additionalQuestions.rename(columns=translatedColumns)

additionalQuestions=additionalQuestions.dropna()

additionalQuestions=additionalQuestions.replace(groupModification)
additionalQuestions
```

### Demographics

```python
#Column renaming:
demographics=demographics.rename(columns={'Edad': 'age', 'GRUPO': 'group','FIRMAS':'attendance','Sexo':'gender','Estudios':'education','Exp Laboral(años)':'experience','Exp en Empresa':'experience in the company','Perfil':'role','N Empleado': 'ID'})

#Numercial columns casting:
demographics[['age', 'attendance','experience','experience in the company']] = demographics[['age', 'attendance','experience','experience in the company']].apply(pd.to_numeric)

# Level names replacement for english translation:
#          Gender translation:
genderTranslation={'hombre':'male','mujer':'female'}
demographics=demographics.replace(genderTranslation)
#          Education translation:
demographics['education']=demographics['education'].fillna('Unspecified')
educationTranslation={'c_medio':'Vocational Training',
              'c_superior':'Advanced Vocational Training',
              'diplomado':'Associate Degree',
              'licenciado':'Bachelor Degree'}
demographics=demographics.replace(educationTranslation)
#          Role translation:
roleTranslation={'desarrollo':'development',
                  'soporte':'support',
                  'sistemas':'systems',
                 'desarrollo/soporte':'development/support',
                 'administracion':'management',
                 'direccion':'director',
                 'Desarrollo':'Development',
                 'Soporte':'Support'}
demographics=demographics.replace(roleTranslation)
groupModification={'G1':'EXP','G2':'CTRL'}
demographics=demographics.replace(groupModification)
#demographics['attendance']=demographics['attendance'].astype('int')
# We specify as exclusion criteria that experimental subjects must attend a minimum of 50% of session
minimumAttendance=(max(demographics['attendance'])/2)
print('Minimum attendance threshold')
print(minimumAttendance)
subjectsToExcludeDueToAttendance=demographics.loc[demographics['attendance']<minimumAttendance]
print("Subjects to be excluded:")
print(subjectsToExcludeDueToAttendance[['ID','group','attendance']])
IDofSubjectsToExcludeDueToAttendance=subjectsToExcludeDueToAttendance['ID']
print("IDs to be excluded")
print(IDofSubjectsToExcludeDueToAttendance)
# We remove the subjects from the different datasets:
# MAAS
print("MAAS exclulsion")
print(" Rows prior to exclusion:"+str(len(maas.index)))
maas=maas.loc[~maas['ID'].isin(IDofSubjectsToExcludeDueToAttendance)]
print(" Rows after exclusion:"+str(len(maas.index)))
# Techno-stress
print("Techno-stress exclulsion")
print(" Rows prior to exclusion:"+str(len(additionalQuestions.index)))
additionalQuestions=additionalQuestions.loc[~additionalQuestions['ID'].isin(IDofSubjectsToExcludeDueToAttendance)]
print(" Rows after exclusion:"+str(len(additionalQuestions.index)))
# Welfare
print("Welfare exclulsion")
print(" Rows prior to exclusion:"+str(len(welfare.index)))
welfares=welfare.loc[~welfare['ID'].isin(IDofSubjectsToExcludeDueToAttendance)]
print(" Rows after exclusion:"+str(len(welfare.index)))
demographics
```

## Analysis with further factors and Data Export for analysis with JASP

```python
maasPRE=maas.loc[maas['MOMENT']=='PRE']
maasPRE['attentionAwarenessPRE']=maasPRE['attention awareness']
maasPOST=maas.loc[maas['MOMENT']== 'POST']
maasPOST['attentionAwarenessPOST']=maasPOST['attention awareness']
maasDataForJASP=pd.merge(maasPRE[['ID','GROUP','attentionAwarenessPRE']],
                                     maasPOST[['ID','attentionAwarenessPOST']],on='ID')
maasDataForJASP=pd.merge(maasDataForJASP,
                                     demographics[['ID','role','Dept']], on='ID')
maasDataForJASP['DeptAsNumeric']=maasDataForJASP['Dept']
maasDataForJASP['DeptAsNumeric']=maasDataForJASP['DeptAsNumeric'].replace({'Development':0,'Support':1})
maasDataForJASP.to_csv('data/maasDataForJASP.csv')
```

```python
additionalQuestionsPre=additionalQuestions.loc[additionalQuestions['MOMENT']=='PRE']
additionalQuestionsPre['TechnoStressPRE']=additionalQuestionsPre['Techno-stress']
additionalQuestionsPost=additionalQuestions.loc[additionalQuestions['MOMENT']=='POST']
additionalQuestionsPost['TechnoStressPOST']=additionalQuestionsPost['Techno-stress']
technoStressDataForJASP=pd.merge(additionalQuestionsPre[['ID','GROUP','TechnoStressPRE']],
                                     additionalQuestionsPost[['ID','TechnoStressPOST']],on='ID')
technoStressDataForJASP=pd.merge(technoStressDataForJASP,demographics[['ID','role','Dept']], on='ID')
technoStressDataForJASP['DeptAsNumeric']=technoStressDataForJASP['Dept']
technoStressDataForJASP['DeptAsNumeric']=technoStressDataForJASP['DeptAsNumeric'].replace({'Development':0,'Support':1})
technoStressDataForJASP.to_csv('data/technoStressDataForJASP.csv')
```
